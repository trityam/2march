```python
# 2,Ans:___
"""The matplotlib.pyplot.scatter() plots serve as a visual tool to explore and analyze the relationships between 
variables, utilizing dots to depict the connection between them"""
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))


plt.scatter(x,y)
plt.title('NUMBER OF POPULATION IN AREA')
plt.xlabel('area')
plt.ylabel('population')
plt.show()
```

```python
#3,ANS:_
""" subplots() function in Python simplifies the creation of multiple subplots Matplotlib within a single 
figure"""
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(x1,y1)

plt.subplot(2,2,3)
plt.plot(x2,y2)

plt.subplot(2,2,4)
plt.plot(x3,y3)
```

```python
#4.Ans_
"""A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths
and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally
or vertically"""
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(2,2,1)
plt.bar(company,profit)

plt.subplot(2,2,2)
plt.barh(company,profit)
```
